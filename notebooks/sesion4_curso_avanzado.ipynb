### Construir un dataframe con valores aleatorios

````python
import random
import string
````

````python
# creamos una lista de letras
columnas = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
````

````python
vals = []
for i in range(5):
    print(random.sample(range(1, 50), 10))
    vals.append(random.sample(range(1, 50), 10))
````

````python
df = DataFrame(vals, columns = columnas)
df
````

````python
# media por filas `axis = 1`
df.mean(axis = 1)
````

````python
# media por clumnas `axis = 0`
df.mean(axis = 0)
````

````python
# std por filas `axis = 1`
df.std(axis = 1)
````

````python
# std por columnas `axis = 0`
df.std(axis = 0)
````

````python
df
````

````python
df.std(axis = 0).round(2)
````

````python
dict(zip(columnas, df.std(axis = 0).round(2).tolist()))
````

A box plot is used to visualize 5 values in a dataset for the selected column(s):

```
    Minimum Value
    First Quartile or 25%
    Median (Second Quartile) or 50%
    Third Quartile or 75%
    Maximum value
```

````python
# filtrar los valores con un umbral
df3 = df2[df2['Entry'] >= 500].reset_index(drop = True)
df3
````
